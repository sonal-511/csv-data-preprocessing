**Code to convert xml to csv**

```
import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET


def xml_to_csv(path):
    xml_list = []
    for xml_file in glob.glob(path + '/*.xml'):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            value = (root.find('filename').text,
                     int(root.find('size')[0].text),
                     int(root.find('size')[1].text),
                     member[0].text,
                     int(member[4][0].text),
                     int(member[4][1].text),
                     int(member[4][2].text),
                     int(member[4][3].text)
                     )
            xml_list.append(value)
    column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df




```

```
image_path = "/content/drive/My Drive/Dataset/dataset_final/annotation"
xml_df = xml_to_csv(image_path)
xml_df.to_csv('dataset_label.csv', index=None)
print('Successfully converted xml to csv.')

```

**Shuffling in groups train test files**

```
path_test="/content/drive/My Drive/Dataset/dataset_final/complete_test.csv"
path_train="/content/drive/My Drive/Dataset/dataset_final/complete_train.csv"
df_train=pd.read_csv(path_train,index_col=0)
df_test=pd.read_csv(path_test,index_col=0)
groups_test = [df_test for _, df_test in df_test.groupby('filename')]
groups_train=[df_train for _, df_train in df_train.groupby('filename')]
random.shuffle(groups_train)
random.shuffle(groups_test)


pd.concat(groups_test).to_csv("/content/drive/My Drive/Dataset/dataset_final/complete_test_shuffled.csv")
pd.concat(groups_train).to_csv("/content/drive/My Drive/Dataset/dataset_final/complete_train_shuffled.csv")
```
